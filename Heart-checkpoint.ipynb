Taken from https://scikit-learn.org/stable/auto_examples/tree/plot_cost_complexity_pruning.html

```python
path = DT.cost_complexity_pruning_path(xtrain, ytrain)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], impurities[:-1], marker='o', drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
plt.savefig('effective_alpha')
```

```python
DTs = []
for ccp_alpha in ccp_alphas:
    clf = tree.DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(xtrain, ytrain)
    DTs.append(clf)
print("Number of nodes in the last tree is: {} with ccp_alpha: {}".format(DTs[-1].tree_.node_count, ccp_alphas[-1]))
```

```python
DTs = DTs[:-1]
ccp_alphas = ccp_alphas[:-1]

node_counts = [clf.tree_.node_count for clf in DTs]
depth = [clf.tree_.max_depth for clf in DTs]
fig, ax = plt.subplots(2, 1)
ax[0].plot(ccp_alphas, node_counts, marker='o', drawstyle="steps-post")
#plt.savefig('Number of nodes vs alpha')
ax[0].set_xlabel("alpha")
ax[0].set_ylabel("number of nodes")
ax[0].set_title("Number of nodes vs alpha")
ax[1].plot(ccp_alphas, depth, marker='o', drawstyle="steps-post")
#plt.savefig('Depth vs alpha')
ax[1].set_xlabel("alpha")
ax[1].set_ylabel("depth of tree")
ax[1].set_title("Depth vs alpha")
fig.tight_layout()
```

```python
train_scores = [clf.score(xtrain, ytrain) for clf in DTs]
test_scores = [clf.score(xtest, ytest) for clf in DTs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
#plt.savefig('Accuracy vs alpha')
```

```python
n_neighbors= []
acc =[]

for i in range(1,40):
    KNN = KNeighborsClassifier(n_neighbors=i)
    KNN.fit(xtrain_std,ytrain)
    ypred['KNN_test'] = KNN.predict(xtest_std)
    acc.append(accuracy_score(ytest , ypred['KNN_test']))
    
    n_neighbors.append(i)
    
    
dk = pd.DataFrame({'acc': acc,'n_neighbors':n_neighbors}) 


plt.plot('n_neighbors','acc', data=dk , label='knn')

plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.legend()
#plt.savefig('acc_n_neighbors_knn')
```

Taken from https://stackabuse.com/k-nearest-neighbors-algorithm-in-python-and-scikit-learn

```python
error = []

# Calculating error for K values between 1 and 40
for i in range(1, 40):
    KNN= KNeighborsClassifier(n_neighbors=i)
    KNN.fit(xtrain_std, ytrain)
    pred_i = KNN.predict(xtest_std)
    error.append(np.mean(pred_i != ytest))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
plt.savefig('Error Rate K Value')
```
